Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Creating the Dataframe

```python
dataset = pd.read_csv('Data_2.csv')
print(dataset)
```

Creating the Distribution for the Salary Variable

```python
plt.hist(dataset['Salary'], color ='blue', edgecolor = 'black', bins= int(35000/1000))
plt.axvline(dataset['Salary'].mean(), color ='red', linewidth = 2, linestyle = 'dashed')
plt.axvline(dataset['Salary'].median(), color ='green', linewidth = 2, linestyle = 'dashed')
```

Computing Salary's Mean

```python
dataset.loc[:,'Salary'].mean()
```

Computing Salary's Min

```python
dataset.loc[:,'Salary'].min()
```

Computing Salary's Max

```python
dataset.loc[:,'Salary'].max()
```

Computing Variance

```python
dataset.loc[:,'Salary'].var()
```

Computing Standard Deviation

```python
dataset.loc[:,'Salary'].std()
```

Computing Skewness of Salary Distribution

```python
dataset.loc[:,'Salary'].skew()
```

Computing Kurtosis of Salary Distribution

```python
dataset.loc[:,'Salary'].kurt()
```

Computing the Mean for both Age and Salary

Computing Skewness for Age and Salary

```python
dataset.skew()
```

Computing Kurtosis for both Age and Salary
