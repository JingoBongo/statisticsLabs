```python
dataset.mean()
```

Computing Medians

```python
median = np.median(dataset['Age'])
print(median)
```

```python
plt.hist(dataset['Salary'], color = 'blue', edgecolor = 'black',
         bins = int(35000/1000))
plt.axvline(dataset['Salary'].mean(), color='red', linestyle='dashed', linewidth=1)
plt.axvline(dataset['Salary'].median(), color='green', linestyle='dashed', linewidth=1)
```
