Importing the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Uploading the Dataset

```python
dataset =pd.read_csv('50_Startups_1.csv')
print(dataset)
```

Creating a variable to count the value frequency

```python
Startups =  dataset['State'].value_counts()
print(Startups)
```

```python
States = ('California', 'New York', 'Florida')
print(States)
```

Building a Bar Chart to show the number of startups in each state in the dataset

```python
mycolors =['hotpink', 'green', 'blue']
plt.bar(States, Startups, color=mycolors)
plt.grid(True)
plt.title('Number of Startups by State', fontsize = 14)
plt.xlabel('States', fontsize = 14)
plt.ylabel('Number of Startups', fontsize = 14)
```

Creating variable amount to sum up the spendings by group of spending (i.e. R&D, Marketing)

```python
Spend =['Reasearch & Development', 'Administration', 'Marketing']
Amount = [dataset.RD.sum(), dataset.Admin.sum(), dataset.Marketing.sum()]
print(Amount)
```

Creating a Bar Chart to show the amount of spendings for startups by type of spending

```python
mycolors =['hotpink', 'green', 'cyan']
plt.bar(Spend, Amount, color=mycolors)
plt.grid(True)
plt.title('Number of Startups by State', fontsize = 14)
plt.xlabel('States', fontsize = 14)
plt.ylabel('Number of Startups', fontsize = 14)
```

To better see relative frequency, or the amount spent for each type of expence as a fraction of the total expences create a Pie Plot 

```python
labels = Spend
explode=(0.1, 0, 0)
mycolors =['hotpink', 'green', 'cyan']
plt.pie(Amount, labels=labels, autopct='%1.1f%%', startangle=90, shadow=True,
        explode=explode, colors=mycolors)
plt.legend(title='Expenditures by type', bbox_to_anchor=(0.9, 0.2),
           bbox_transform=plt.gcf().transFigure)

plt.show()
```
